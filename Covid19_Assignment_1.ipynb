# 3. Load global data from csv file

```python
import pandas as pd

df = pd.read_csv('time_series_covid19_confirmed_global.csv')
df.head(3)
```

## 4. Count and display the total number of NaN values if any

```python
# Total number of NaN values across the whole dataframe
df.isnull().sum().sum()
```

## 5. Replace the NaN values with the text 'not available'
- Show some rows with updated text

```python
df = df.fillna('not available')
df.head()
```

# 6. Load only China's data

```python
china = df[df['Country/Region'] == 'China']
china.head(3)
```

## 7. Update the index
- If the indexes dont start from 0, then reset it to start from 0.
- You may need a new dataframe.

```python
china = china.reset_index(drop=True)
china.head()
```

## 8. Discard Country/region, Lat and Long columns
- Creat a new dataframe

```python
china_cases = china.drop(columns=['Country/Region', 'Lat', 'Long'])
china_cases.head()
```

# 9. Transpose the table. Provinces as columns and Cases as rows

```python
# Provinces become the columns and the daily cases become the rows
china_t = china_cases.T
china_t.columns = china_t.iloc[0]                 # province names as column headers
china_t = china_t.drop(china_t.index[0]).reset_index(drop=True)
china_t.columns.name = None
china_t.head()
```
